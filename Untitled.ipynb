# Regression Setting

### Setting 1: Linear regression correctly specified

```python
import numpy as np
from scipy.stats import multivariate_normal as mvn
from scipy.stats import multinomial, norm, binom, laplace
from tqdm import tqdm
import matplotlib.pyplot as plt
```

The true model is $y=X\beta_*$. From this truth, we observe noisy $(x,y)$ as $y=X\beta^* + \epsilon$. 

At iteration $t$, we observe $\{(X_j^{(t)}, y_j^{(t)}) : j\in [m_t]\}$ coming from a mixture of the true model and the last fitted model. The variance of the noise is a nuisance parameter and the working model assumes it is standard normal (always - no estimation of this variance), since the target is only prediction accuracy. Thus,

$$y_j^{(t)} = \begin{cases}
    X_j^{(t)}\beta_* + \epsilon & \text{ with probability } \alpha_t \\
    X_j^{(t)}\hat{\beta}_{t-1} + \tilde{\epsilon} &\text{ with probability } 1-\alpha_t
\end{cases}$$

where $\hat{\beta}_{t-1}$ is the estimated value of the coefficients at the previous iteration. At $t=0$, we only observe from true model.

For simulations, assume $\epsilon \sim N(0, 1.5)$, while the working model, $\tilde{\epsilon}\sim N(0,1)$ and these noise terms are independent of all other processes.

```python
def generate_normal(beta, X_mu, X_Sigma, noise_sigma, n):
    d = len(beta) -1
    X = mvn.rvs(mean=X_mu, cov=X_Sigma, size=n).reshape((n,d))
    X = np.hstack([np.ones(n).reshape((n,1)), X])
    eps = mvn.rvs(mean=[0], cov=[[noise_sigma**2]], size=n)
    Y = X @ beta + eps
    return (X, Y)

def estimate(X, Y):
    A = X.T @ X
    b = X.T @ Y
    return np.linalg.solve(A, b)

def distance(beta1, beta2, Sigma=None):
    if Sigma is None:
        return np.linalg.norm(beta1 - beta2)
    else:
        beta = beta1 - beta2
        return np.sqrt(beta.T @ Sigma @ beta)
    
def generate_mixture(beta, X_mu, X_Sigma, noise_sigma, beta_working, noise_working, alpha, n):
    d = len(beta) -1
    X = mvn.rvs(mean=X_mu, cov=X_Sigma, size=n).reshape((n,d))
    X = np.hstack([np.ones(n).reshape((n,1)), X])
    eps1 = mvn.rvs(mean=[0], cov=[[noise_sigma**2]], size=n)
    eps2 = mvn.rvs(mean=[0], cov=[[noise_working**2]], size=n)
    Y1 = X @ beta + eps1 # true model
    Y2 = X @ beta_working + eps2 # working model
    group = binom.rvs(n=1, p=alpha, size=n).flatten() # 1 is true model
    Y = Y1 * group + Y2 * (1-group)
    return (X, Y)
```

```python
d = 1
beta0 = np.random.normal(0, 1, d+1)
print(f'beta = {beta0}')
X_mu = np.zeros(d)
X_Sigma = np.eye(d)
noise_sigma = 1.5

beta_working = -beta0

X, Y = generate_mixture(beta0, X_mu, X_Sigma, noise_sigma, beta_working, 1.0, 0.5, 1000)
```

```python
beta_hat = estimate(X, Y)
beta_hat

fig, ax = plt.subplots(figsize=(5,3))
plt.scatter(X[:,1], Y, color='gray', alpha=0.3)
plt.axline((0, beta0[0]), slope=beta0[1], label='model1', color='red', linestyle='dashed')
plt.axline((0, beta_working[0]), slope=beta_working[1], label='model2', color='blue', linestyle='dashed')
plt.axline((0, beta_hat[0]), slope=beta_hat[1], label='estimate', color='orange', linestyle='dashed')
plt.legend()
plt.title('Mixture of regression example')
plt.show()
```

We track $E\left[\Vert\hat{\beta}_t - \beta_*\Vert_2^2\right]$ to measure performance.

Moreover, we focus on the case there is no accumulation. The sample size increases linearly $m_t=m_0 t$ and proportion of data coming from truth decreasing $\alpha_t = \alpha_0 (1/t)^\beta$.

```python
T = 500
rep = 50
M = 20
rates = [0.0, 0.33, 0.5, 1.0, 2.0]
alpha0 = 0.8

dist_mat = np.zeros((len(rates), rep, T))

for j in range(len(rates)):
    alpha = alpha0
    rate = rates[j]
    for i in tqdm(range(rep)):
        X, Y = generate_normal(beta0, X_mu, X_Sigma, noise_sigma, M)
        beta_hat = estimate(X, Y)
        dist_mat[j, i, 0] = np.linalg.norm(beta0 - beta_hat)


        for t in range(T-1):
            alpha = alpha0 / (t+1)**rate
            X, Y = generate_mixture(beta0, X_mu, X_Sigma, noise_sigma, beta_hat, 1.0, alpha, M*(t+2))
            beta_hat = estimate(X, Y)
            dist_mat[j, i, t+1] = np.linalg.norm(beta0 - beta_hat)
```

```python
dist_mean = dist_mat.mean(axis=1)
dist_sd = dist_mat.std(axis=1)
M_list = np.sqrt(M * np.arange(1, T+1))

fig, ax = plt.subplots(figsize=(15, 4), ncols=5)

for i in range(len(rates)):
    ax[i].plot(range(T), dist_mean[i], label=rates[i])
    ax[i].fill_between(range(T), dist_mean[i] - 1.96*dist_sd[i]/M_list, dist_mean[i] + 1.96*dist_sd[i]/M_list, alpha=0.2)
    ax[i].set_title(f'alpha_t={alpha0}/t^{rates[i]}')
    ax[i].set_xlabel('T')
ax[0].set_ylabel('sq. error')
plt.suptitle(f'Mis-Specified, no accumulation - m increasing, alpha decreasing')
plt.tight_layout()
plt.show()
```
